```python
from sklearn.preprocessing import StandardScaler


# Separate features and target
X = df.drop('letter', axis=1).values
y = df['letter'].values

# Normalize the feature data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#Model implementation

import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input


label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)


# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_encoded, test_size=0.2, random_state=42)

# Build the ANN model
input_dim = X_train.shape[1]

model = Sequential()
model.add(Input(shape=(input_dim,)))  # Input layer
model.add(Dense(128, activation='relu'))  # First hidden layer
model.add(Dense(64, activation='relu'))   # Second hidden layer
model.add(Dense(len(set(y)), activation='softmax'))  # Output layer

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2)
```

```python
#Predictions

# Predictions on the test set
predictions = model.predict(X_test)
predicted_classes = predictions.argmax(axis=1)
```

```python
#parameter Tuning
from sklearn.model_selection import ParameterGrid
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input

# Create parameter grid
param_grid = {
    'batch_size': [16, 32],
    'epochs': [50, 100],
    'optimizer': ['adam', 'sgd'],
    'neurons': [64, 128]
}

# Loop over parameter grid
best_score = 0
best_params = {}

for params in ParameterGrid(param_grid):
    # Create model
    model = Sequential()
    model.add(Input(shape=(X_train.shape[1],)))  # Input layer
    model.add(Dense(params['neurons'], activation='relu'))  # Hidden layer
    model.add(Dense(64, activation='relu'))      # Second hidden layer
    model.add(Dense(len(set(y)), activation='softmax'))  # Output layer
    
    model.compile(optimizer=params['optimizer'], loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    
    # Train the model
    model.fit(X_train, y_train, epochs=params['epochs'], batch_size=params['batch_size'], verbose=0)
    
    # Evaluate the model
    score = model.evaluate(X_test, y_test, verbose=0)
    
    # Check if the current model is better
    if score[0] < best_score or best_score == 0:
        best_score = score[0]
        best_params = params

print(f"Best parameters: {best_params}")
print(f"Best score: {best_score}")
```

```python
#Evaluation


from sklearn.metrics import classification_report, accuracy_score

# Evaluate on test set
test_accuracy = accuracy_score(y_test, predicted_classes)
print(f'Test Accuracy: {test_accuracy}')

# Detailed classification report
print(classification_report(y_test, predicted_classes))
```
